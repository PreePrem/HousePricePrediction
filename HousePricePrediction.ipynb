```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
data = pd.read_csv('/content/Housing.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
print(data.furnishingstatus.value_counts())
print(data.mainroad.value_counts())
print(data.guestroom.value_counts())
print(data.basement.value_counts())
print(data.hotwaterheating.value_counts())
print(data.airconditioning.value_counts())
print(data.prefarea.value_counts())
```

Replacing all the strings into numbers

```python
data.replace({'furnishingstatus':{'unfurnished':0,'semi-furnished':1,'furnished':2}}, inplace=True)
data.replace({'mainroad':{'yes':0,'no':1}}, inplace=True)
data.replace({'guestroom':{'yes':0,'no':1}}, inplace=True)
data.replace({'basement':{'yes':0,'no':1}}, inplace=True)
data.replace({'hotwaterheating':{'yes':0,'no':1}}, inplace=True)
data.replace({'airconditioning':{'yes':0,'no':1}}, inplace=True)
data.replace({'prefarea':{'yes':0,'no':1}}, inplace=True)
```

data split

```python
x = data.drop(columns='price',axis=1)
y = data['price']
x
```

data split

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
y_train
```

Model Training

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

Model Evaluation

```python
x_train_pred = model.predict(x_train)
train_data = accuracy_score(x_train_pred, y_train)
train_data
```

```python
x_test_pred = model.predict(x_test)
test_data = accuracy_score(x_test_pred, y_test)
test_data
```

model2 training random forest

```python
m2 = RandomForestClassifier()
```

```python
m2.fit(x_train, y_train)
```

```python
pred_rf = m2.predict(x_train)
```

```python
accu_score = accuracy_score(pred_rf, y_train)
accu_score
```

```python
pred = m2.predict(x_test)
accu_score = accuracy_score(pred, y_test)
accu_score
```

model 3

```python
m3 = XGBRegressor()
```

```python
m3.fit(x_train, y_train)
```

```python
trainpred = m3.predict(x_train)
```

```python
s1 = metrics.r2_score(y_train, trainpred)
s1
```

```python
testpred = m3.predict(x_test)
```

```python
s2 = metrics.r2_score(y_test, testpred)
s2
```
